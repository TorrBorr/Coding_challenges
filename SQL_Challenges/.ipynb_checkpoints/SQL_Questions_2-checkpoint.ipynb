Which team scored the most points when playing at home?

```python
pd.read_sql_query("""SELECT team_long_name, SUM(home_team_goal) as total_points
                     FROM Match as M 
                     JOIN Team as T
                     ON M.home_team_api_id = T.team_api_id
                     GROUP BY T.team_long_name
                     ORDER BY total_points DESC
                     LIMIT 1
                     ;""", conn)
```

Did this team also score the most points when playing away?

```python
pd.read_sql_query("""SELECT team_long_name, SUM(away_team_goal) as total_points
                     FROM Match as M
                     JOIN Team as T
                     ON M.away_team_api_id = T.team_api_id
                     GROUP BY team_long_name
                     ORDER BY total_points DESC
                     LIMIT 5
                     ;""", conn)
```

What percentage of players prefer their left or right foot? Hint: Calculate either the right or left foot, whichever is easier based on how you setup the problem.

```python
pd.read_sql_query("""SELECT preferred_foot, COUNT(preferred_foot) as num
                     FROM Player_Attributes
                     GROUP BY preferred_foot;""", conn)
```

```python
pd.read_sql_query("""WITH foot_table AS (SELECT preferred_foot, 
                                         COUNT(preferred_foot) as num
                                         FROM Player_Attributes
                                         GROUP BY preferred_foot)
                     SELECT preferred_foot, num,  100*num/(SUM(num)) as percentage
                     FROM foot_table
                     ;""", conn)
```
