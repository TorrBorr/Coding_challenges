Cluster items based on user co-purchase history. That is, create clusters of products that have the highest probability of being bought together. The goal of this is to replace the old/manually created categories with these new ones. Each item can belong to just one cluster.

```python
df_hist_items= df_hist.set_index(['user_id'])['id_list'].apply(pd.Series).stack()
df_hist_items = df_hist_items.reset_index()
df_hist_items.columns = ["user","Number_purchased","Item_id"]
df_hist_items["Item_id"] = df_hist_items["Item_id"].astype(str).astype(int)
df_hist_items = df_hist_items.groupby(['user','Item_id']).count().reset_index()
df_hist_items.head(5)
```

```python
df_hist_items.head(20)
```

```python
df_hist_name = pd.merge(df_hist_items, df_item, on = 'Item_id') #\
                 #.drop(['prod_id_num','Item_id'],axis =1)\
                 #.groupby('user')
```

```python
df_hist_name.head(20)
```

```python
purchase_table = df_hist_name.pivot(index = 'user', columns = "Item_name", values = "Number_purchased")
```

```python
purchase_table = purchase_table.replace(np.NaN,0)

purchase_table.head()
```
